```python
import numpy as np
import cv2
import matplotlib.image as matimg
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# OpenCV - BGR - numpy array
img_cv = cv2.imread('./data/flying_bird.jpg')

# Matplotlib - RGB - numpy array
img_mat = matimg.imread('./data/flying_bird.jpg')

# PIL - RGB - Object
img_pil = Image.open('./data/flying_bird.jpg')
```

```python
# OpenCV
print(img_cv)
```

```python
# matplotlib
print(img_mat)
```

```python
img_pil
```

```python
img_pil_arr = np.array(img_pil)
img_pil_arr
```

```python
# display image using OpenCV
cv2.imshow('OpenCV BGR', img_cv)
cv2.imshow('Matplotlib RGB', img_mat)
cv2.imshow('PIL RGB', img_pil_arr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# display image using Matplotlib
plt.figure(figsize=(15,8))
plt.subplot(1, 3, 1)
plt.imshow(img_cv)
plt.axis('off')
plt.title('OpenCV BGR')

plt.subplot(1, 3, 2)
plt.imshow(img_mat)
plt.axis('off')
plt.title('Matplotlib RGB')

plt.subplot(1, 3, 3)
plt.imshow(img_pil_arr)
plt.axis('off')
plt.title('PIL RGB')

plt.show()
```

```python
img_cv.shape
```

```python
# numpy slicing
# BGR
b = img_cv[:,:,0]
g = img_cv[:,:,1]
r = img_cv[:,:,2]
```

```python
cv2.imshow('blue channel', b)
cv2.imshow('green channel', g)
cv2.imshow('red channel', r)

cv2.imshow('BGR', img_cv)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# OpenCV split channesl
b, g, r, = cv2.split(img_cv)
```

```python
cv2.imshow('blue channel', b)
cv2.imshow('green channel', g)
cv2.imshow('red channel', r)

cv2.imshow('BGR', img_cv)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# converting color
# bgr to rgb
img_cv_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_cv_rgb)
plt.axis('off')
plt.show()
```

```python
img_gray = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('grayscale', img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
